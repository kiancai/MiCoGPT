- 安装 qiime2

  ```shell
  CONDA_SUBDIR=osx-64 conda env create \
    --name qiime2-amp-202510 \
    --file https://raw.githubusercontent.com/qiime2/distributions/refs/heads/dev/2025.10/amplicon/released/qiime2-amplicon-macos-latest-conda.yml
  conda activate qiime2-amp-202510
  conda config --env --set subdir osx-64
  ```

- 将 conda 环境注册为 Jupyter 内核

  ```shell
  conda install python=3.10 ipykernel ipywidgets
  python -m ipykernel install --user --name=qiime2-amp-202510 --display-name "Python (qiime2-amp-202510)"
  ```

- 下载 SILVA 数据库

```shell
  wget https://data.qiime2.org/2024.10/common/silva-138-99-seqs.qza
  wget https://data.qiime2.org/2024.10/common/silva-138-99-tax.qza
```


- 导出 SILVA 数据库

```shell
  qiime tools export \
    --input-path silva-138-99-seqs.qza \
    --output-path silva-138-99-seqs-exported

  qiime tools export \
    --input-path silva-138-99-tax.qza \
    --output-path silva-138-99-tax-exported
```
